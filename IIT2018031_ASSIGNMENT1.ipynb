```
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread("/content/index.jpeg")
```

```
# 1)
shape = img.shape
print(shape)
```

```
2 #a)
grayscale = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(grayscale)
```

```
#b)
import numpy as np
# pixels_of_value_6 = []
six_in_grayscale = np.argwhere(grayscale == 6)
six_in_img = np.argwhere(img == 6)
print("six_in_grayscale:",six_in_grayscale)
print("six_in_img",six_in_img)
```

```
# c)
minElement_in_grayscale = np.amin(grayscale)
minElement_in_img = np.amin(img)
print("minElement_in_grayscale",minElement_in_grayscale)
print("minElement_in_img",minElement_in_img)
```

```
# d)
result = np.where(grayscale == np.amin(grayscale))
cpy = np.copy(grayscale)
for point in result:
  for i in range(max(0,point[0]-15),min(point[0]+15,shape[0])):
    for j in range(max(0,point[1]-15),min(point[1]+15,shape[1])):
      cpy[i][j]=255
cv2_imshow(cpy)
```

```
# e)
cpy = np.copy(grayscale)
point = [int(shape[0]/2),int(shape[1]/2)]
for i in range(max(0,point[0]-15),min(point[0]+15,shape[0])):
  for j in range(max(0,point[1]-15),min(point[1]+15,shape[1])):
    cpy[i][j]=100
cv2_imshow(cpy)
```

```
# 3
print(np.mean(img[:,:,0]),
np.mean(img[:,:,1]),
np.mean(img[:,:,2]))
img[:,:,0] -= np.mean(img[:,:,0]).astype(int)
img[:,:,1] -= np.mean(img[:,:,0]).astype(int)
img[:,:,2] -= np.mean(img[:,:,0]).astype(int)
cv2_imshow(img)
```

```
# 4) a)
img = cv2.imread("/content/index.jpeg")
M = cv2.getRotationMatrix2D((shape[0] / 2, shape[1] / 2), 45, 1) 
res = cv2.warpAffine(img, M, (shape[0], shape[1])) 
cv2_imshow(res)
```

```
# b)
res = cv2.resize(img, (int(shape[1] / 2), int(shape[0] / 2)))
cv2_imshow(res) 
```

```
# c)
M = np.float32([[1, 0, 10], [0, 1, 10]]) 
res = cv2.warpAffine(img, M, (shape[1], shape[0])) 
cv2_imshow(res)
```

```
# d)
edges = cv2.Canny(img, 100, 200) 
cv2_imshow(edges)
```
